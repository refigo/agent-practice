```python
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()
client = OpenAI()
```

```python
system_prompt = """
You are a movie recommendation expert who speaks Korean.

Your responsibilities:
1. Remember the user's favorite genres throughout the conversation.
2. Remember movies the user has already watched.
3. When recommending movies, NEVER recommend movies the user has already watched.
4. Provide personalized recommendations based on the user's preferences and watch history.
5. When asked, accurately recall the user's favorite genres and watched movies from the conversation.
"""

messages = [{"role": "system", "content": system_prompt}]
```

```python
def chat(user_message):
    print(f"User: {user_message}\n")
    messages.append({"role": "user", "content": user_message})

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
    )

    reply = response.choices[0].message.content
    messages.append({"role": "assistant", "content": reply})

    print(f"AI: {reply}")
    return reply
```

```python
chat("나는 SF 영화를 좋아해")
```
